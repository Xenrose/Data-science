## 1. 기본식 (기본 폰트 지정, X축/Y축/Title 각종 설정)

```python
# 기본 폰트 지정

import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결


# 사용 가능한 폰트 확인
import matplotlib.font_manager as fm
fm.fontManager.ttflist 
[f.name for f in fm.fontManager.ttflist]
```

```python
# x축 y축 Data 지정
x = [1, 2, 3]
y = [2, 4, 8]



# title 설정
plt.title('Title_name')
# title 메서드에 fontdict 파라미터를 사용하여 개별 폰트 설정이 가능하다.
# plt.title('Title_name', fontdict={'family':'HYGungSo-Bold', 'size':20}) # 개별 폰트 설정


# X, Y축 Label 설정
plt.xlabel('X_label', color='red', loc='right')  
plt.ylabel('Y_label', color='#00aa00', loc='top')


# X, Y축 데이터 범주 설정
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])


plt.plot(x, y, label='범례')  # show() 메서드를 사용하지 않아도 이것만 사용해도 출력되긴 함.
plt.legend(loc=(0.75, 0.02)) # x축, y축 (0~1 사이) // 선 색상 설명을 표기해줌
plt.show() 
```

____
## 2. 기본 스타일 설정

```python
# 마커 설정
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')

# marker = o / v / X  // 마커 모양
# markersize = 00 // 마커의 크기
# markeredgecolor = 'red' // 마커의 테두리 색
# markerfacecolor = 'yellow' // 마커의 색
```

```python
# 마커에 Text 입력

plt.plot(x, y, marker='o')

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx] + 0.3, txt, ha='center', color='blue')
    # text(x, y)
    # ha = 'center' // 중앙 정렬
```

```python
# 선 설정

plt.subplot(2,2,1)
plt.plot(x, y, color='g', marker='o', linestyle='None')

plt.subplot(2,2,2)
plt.plot(x, y, linewidth=5, linestyle='--', color='pink')

plt.show()

# linewidth = 5 // 선 두께
# linestyle = 'None'(선 없애기) , : , -- , -. , - // 선 모양
# color='pink' or '#ff0000' or 'b' // 선 색
```

```python
# 포맷
plt.subplot(2,2,1)
plt.plot(x, y, 'ro--') # color, marker, linestyle

plt.subplot(2,2,2)
plt.plot(x, y, 'bv:')

plt.subplot(2,2,3)
plt.plot(x, y, 'go')

plt.subplot(2,2,4)
plt.plot(x, y, color='g', marker='o', linestyle='None')

plt.show()
```

```python
# 투명도

plt.subplot(2,2,1)
plt.plot(x, y, 'ro--', alpha=0)

plt.subplot(2,2,2)
plt.plot(x, y, 'ro--', alpha=0.25)

plt.subplot(2,2,3)
plt.plot(x, y, 'ro--', alpha=0.5)

plt.subplot(2,2,4)
plt.plot(x, y, 'ro--', alpha=1)

plt.show()
```

```python
# 그래프 크기
plt.figure(figsize=(10,10), dpi=150) #dpi = dots per inch
plt.subplot(2,2,1)
plt.plot(x, y, 'ro--', alpha=0)

plt.subplot(2,2,2)
plt.plot(x, y, 'ro--', alpha=0.25)

plt.subplot(2,2,3)
plt.plot(x, y, 'ro--', alpha=0.5)

plt.subplot(2,2,4)
plt.plot(x, y, 'ro--', alpha=1)

plt.show()
```

```python
# 배경색

plt.figure(figsize=(10,10), facecolor='green')


plt.subplot(2,2,1)
plt.plot(x, y, 'ro--', alpha=0)

plt.subplot(2,2,2)
plt.plot(x, y, 'ro--', alpha=0.25)

plt.subplot(2,2,3)
plt.plot(x, y, 'ro--', alpha=0.5)

plt.subplot(2,2,4)
plt.plot(x, y, 'ro--', alpha=1)

plt.show()
```
